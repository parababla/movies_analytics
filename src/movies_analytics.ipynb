### Анализ файла: movies.csv
Общая информация \
Файл movies.csv содержит основную информацию о фильмах: их идентификаторы, названия и жанры. Это фундамент нашего анализа — без понимания того, какие фильмы у нас есть, невозможно анализировать оценки или теги.

__________________
Конструктор класса Movies \
Конструктор принимает путь к файлу movies.csv и сохраняет его в self.path. Основная инициализация происходит через вызов метода read_file(), который загружает и обрабатывает первые 1000 строк файла. Результат сохраняется в self.data как список списков, где каждый внутренний список содержит: целочисленный movieId, строку title и список жанров. Метод read_file() использует str_handler() для парсинга каждой строки CSV с учетом кавычек и разделителей, а также extract_genres() для преобразования строки жанров в список. Конструктор не выполняет дополнительных проверок существования файла, оставляя это на обработку исключений в методах чтения файла.

```python
%timeit

movies = Movies(PATH_TO_CSV_MOVIES)
```

__________________
Распределение фильмов по годам выпуска (dist_by_release) \
Метод использует extract_year() для получения всех годов выпуска, затем применяет Counter для подсчета количества фильмов в каждом году. Результат сортируется по убыванию количества фильмов с помощью sorted() и lambda-функции, возвращающей словарь.

Извлечение года из названия фильма (extract_year) \
Метод использует регулярное выражение для поиска года в круглых скобках в названии фильма. Шаблон r'' ищет 4 цифры внутри круглых скобок, игнорируя пробелы. Все найденные годы возвращаются как список целых чисел.

```python
%timeit
years_distribution = movies.dist_by_release()

print("Распределение фильмов по годам выпуска:")
for year, count in years_distribution.items():
    print(f"  {year}: {count} фильмов")
```

______________
Распределение фильмов по жанрам (dist_by_genres) \
Метод собирает все жанры из всех фильмов в один список, используя extend() для добавления жанров каждого фильма. Затем применяется Counter для подсчета частоты каждого жанра. Результат сортируется по убыванию частоты и возвращается как словарь.

```python
%timeit

genres_distribution = movies.dist_by_genres()

print("Распределение по жанрам:")
for genre, count in genres_distribution.items():
    print(f"  {genre}: {count} фильмов")
```

____________________
Фильмы с наибольшим количеством жанров (most_genres) \
Для каждого фильма вычисляется количество жанров (исключая пустые строки). Создается список кортежей (название, количество_жанров). Список сортируется по убыванию количества жанров с помощью sort() и lambda-функции. Возвращаются первые n элементов как словарь.

```python
%timeit

most_genres_films = movies.most_genres(5)

print("Фильмы с наибольшим количеством жанров:")
for title, genre_count in most_genres_films.items():
    print(f"  '{title}': {genre_count} жанров")
```

### Анализ файла: tags.csv
Общая информация \
Файл tags.csv содержит пользовательские теги — свободные текстовые метки, которые зрители присваивают фильмам. Это неформальные, субъективные описания, которые часто раскрывают личное восприятие фильма.

___________________
Конструктор класса Tags \
Конструктор принимает путь к файлу tags.csv и проверяет его существование через os.path.exists(). При успешной проверке сохраняет путь в self.path и сразу инициализирует атрибут self.tags, вызывая цепочку методов: read_file() загружает первые 1000 строк файла, затем select_tags() извлекает теги из третьей колонки (индекс 2) каждой строки. Все теги сохраняются как список строк в self.tags. Если файл не найден, выбрасывается FileNotFoundError с информацией о проблемном пути. Конструктор обеспечивает немедленную доступность тегов для всех аналитических методов класса.

```python
%timeit

tags = Tags(PATH_TO_CSV_TAGS)
```

_______________
Теги с наибольшим количеством слов (most_words) \
Метод использует регулярное выражение re.split(r'[ /:]+', tag) для разделения тега на слова по пробелам, слэшам и двоеточиям. Для каждого тега вычисляется количество слов, создается словарь "тег: количество_слов". Словарь сортируется по убыванию количества слов, а при равных значениях - по алфавиту тегов. Возвращаются первые n элементов.

```python
%timeit

most_words_tags = tags.most_words(5)

print("Самые многословные теги:")
for tag, word_count in most_words_tags.items():
    print(f"  '{tag}': {word_count} слов")
```

________________
Самые длинные теги по символам (longest) \
Метод удаляет дубликаты тегов с помощью set(), преобразует обратно в список, сортирует по убыванию длины тега (используя len() в lambda-функции), а при равной длине - по алфавиту. Возвращаются первые n элементов.

```python
%timeit

longest_tags = tags.longest(5)

print("Самые длинные теги:")
for tag in longest_tags:
    print(f"  '{tag}' ({len(tag)} символов)")
```

___________
Пересечение самых многословных и самых длинных тегов (most_words_and_longest) \
Метод находит пересечение множеств тегов, полученных из most_words(n) и longest(n). Результат сортируется по алфавиту и возвращается как список.

```python
%timeit

most_words_and_longest_tags = tags.most_words_and_longest(5)

print("Самые многословные и длинные теги:")
for tag in most_words_and_longest_tags:
    print(f"  '{tag}'")
```

__________
Самые популярные теги (most_popular) \
Метод использует Counter для подсчета частоты каждого тега, сортирует результат по убыванию частоты, а при равных частотах - по алфавиту тегов. Возвращаются первые n элементов как словарь.

```python
%timeit

popular_tags = tags.most_popular(10)

print("Самые популярные теги:")
for tag, count in popular_tags.items():
    print(f"  '{tag}': используется {count} раз")
```

____________
Теги, содержащие заданное слово (tags_with) \
Метод проходит по всем тегам, проверяет содержит ли тег (в нижнем регистре) заданное слово (также в нижнем регистре). Уникальные теги собираются в set для удаления дубликатов, затем преобразуются в отсортированный по алфавиту список.

```python
%timeit

search_words = ['love', 'action', 'comedy']

for word in search_words:
    matching_tags = tags.tags_with(word)
    print(f"\nТеги, содержащие '{word}':")
    for tag in matching_tags:
        print(f" - {tag}")
```

### Анализ файла: ratings.csv
Общая информация \
Файл ratings.csv — это сердце нашей аналитики. Здесь содержатся оценки, которые реальные пользователи ставят фильмам. Каждая оценка — это маленькое свидетельство того, как фильм повлиял на конкретного человека.

____________
Конструктор класса Ratings \
Конструктор принимает путь к файлу ratings.csv и сохраняет его в self.path. Основная инициализация происходит через вызов read_file(), который загружает первые 1000 строк данных. Метод read_file() использует str_handler() для преобразования каждой строки в список с правильными типами данных: userId и movieId преобразуются в int, rating - в float, timestamp - в int. Обработанные данные сохраняются в self.data как список списков. Конструктор не содержит внутренних классов - они создаются динамически при вызове соответствующих методов через вложенные классы Movies и Users, которые наследуют данные из внешнего экземпляра Ratings.

```python
%timeit

ratings = Ratings(PATH_TO_CSV_RATINGS)
```

_________________
Распределение оценок по годам (dist_by_year) \
Метод извлекает годы из всех timestamp с помощью extract_year_from_timestamps(), подсчитывает количество оценок в каждом году с помощью Counter и сортирует результат по возрастанию года (ключ lambda item: item[0]).

Извлечение года из timestamp (extract_year_from_timestamps) \
Метод преобразует Unix timestamp в объект datetime с помощью datetime.datetime.fromtimestamp(), затем извлекает год с помощью .year.

```python
%timeit

ratings_by_year = ratings.Movies(ratings, PATH_TO_CSV_MOVIES).dist_by_year()

print("Распределение оценок по годам:")
for year, count in sorted(ratings_by_year.items()):
    print(f" {year} год: {count} оценок")
```

_______________
Распределение по рейтингам (dist_by_rating) \
Метод собирает все значения оценок из объединенной таблицы, подсчитывает их частоту с помощью Counter и сортирует по возрастанию рейтинга.

Объединение таблиц (tables_union) \
Метод создает словарь movieId -> [title, genres] из данных о фильмах, затем для каждой оценки из ratings добавляет соответствующую информацию о фильме. Результатом является объединенная таблица [userId, movieId, rating, timestamp, title, genres].

```python
%timeit

rating_distribution = ratings.Movies(ratings, PATH_TO_CSV_MOVIES).dist_by_rating()

print("Распределение оценок:")
for rating, count in sorted(rating_distribution.items()):
    print(f"  Оценка {rating}: {count} раз")
```

___________
Фильмы с наибольшим количеством оценок (top_by_num_of_ratings) \
Метод извлекает названия фильмов из объединенной таблицы, подсчитывает количество оценок для каждого фильма с помощью Counter, сортирует по убыванию количества оценок и возвращает первые n элементов как словарь.

```python
%timeit

movies_obj = ratings.Movies(ratings, PATH_TO_CSV_MOVIES)
most_rated = movies_obj.top_by_num_of_ratings(10)

print("Топ фильмов по количеству оценок:")
for title, count in most_rated.items():
    print(f"  '{title}': {count} оценок")
```

______________
Топ фильмов по среднему/медианному рейтингу (top_by_ratings) \
Метод группирует оценки по названиям фильмов в словарь "название: [оценки]". Затем для каждого фильма вычисляется либо среднее (get_average), либо медиана (get_median). Результаты сортируются по убыванию метрики и обрезаются до n элементов.

```python
%timeit

top_by_avg = movies_obj.top_by_ratings(10, metric='average')

print("Фильмы с самыми высокими средними оценками:")
for title, avg_rating in top_by_avg.items():
    print(f"  '{title}': {avg_rating}/5")
```

```python
%timeit

top_by_median = movies_obj.top_by_ratings(10, metric='median')

print("Фильмы с самыми высокими медианными оценками:")
for title, median_rating in top_by_median.items():
    print(f"  '{title}': {median_rating}/5")
```

_______________
Самые противоречивые фильмы (top_controversial) \
Метод группирует оценки по фильмам, вычисляет дисперсию для каждого фильма с помощью get_variance(), сортирует по убыванию дисперсии и возвращает первые n элементов.

Вычисление дисперсии (get_variance) \
Метод сначала вычисляет среднее значение списка, затем находит сумму квадратов отклонений каждого элемента от среднего, делит на количество элементов. Результат округляется до 2 знаков после запятой.

```python
%timeit

controversial_films = movies_obj.top_controversial(10)

print("Самые противоречивые фильмы:")
for title, variance in controversial_films.items():
    print(f"  '{title}': дисперсия = {variance}")
```

______________
Распределение пользователей по количеству оценок (dist_by_users) \
Метод использует Counter для подсчета количества оценок, оставленных каждым пользователем. UserId извлекается из первого столбца объединенной таблицы.

```python
%timeit

movies_obj = ratings.Movies(ratings, PATH_TO_CSV_MOVIES)

users_obj = ratings.Users(movies_obj) 

active_users = users_obj.dist_by_users()
top_users = sorted(active_users.items(), key=lambda x: x[1], reverse=True)[:5]

print("Самые активные пользователи:")
for user_id, rating_count in top_users:
    print(f"  Пользователь {user_id}: {rating_count} оценок")
```

_______________
Распределение пользователей по средней/медианной оценке (dist_by_metric) \
Метод группирует оценки по пользователям в словарь "userId: [оценки]", затем для каждого пользователя вычисляет либо среднее, либо медиану. В отличие от аналогичного метода для фильмов, здесь нет ограничения по n элементов.

```python
%timeit

users_average_ratings = users_obj.dist_by_metric(metric='average')
users_median_ratings = users_obj.dist_by_metric(metric='median')

print("Топ пользователей по средней оценке:")
sorted_average = sorted(users_average_ratings.items(), key=lambda x: x[1], reverse=True)[:5]
for user_id, avg_rating in sorted_average:
    print(f" Пользователь {user_id}: средняя оценка {avg_rating:.2f}")

print(f"\nТоп пользователей по медианной оценке:")
sorted_median = sorted(users_median_ratings.items(), key=lambda x: x[1], reverse=True)[:5]
for user_id, median_rating in sorted_median:
    print(f" Пользователь {user_id}: медианная оценка {median_rating:.2f}")
```

_____________
Самые противоречивые пользователи (top_controversial) \
Метод группирует оценки по пользователям, вычисляет дисперсию оценок для каждого пользователя с помощью get_variance(), сортирует по убыванию дисперсии и возвращает первые n элементов.

```python
%timeit

most_controversial_users = users_obj.top_controversial(10)

print("Топ самых противоречивых пользователей (по дисперсии оценок):")
for user_id, variance in most_controversial_users.items():
    print(f" Пользователь {user_id}: дисперсия оценок {variance:.2f}")
```

### Анализ файла: links.csv
Общая информация \
Файл links.csv — наш мост во внешний мир киноданных. Он содержит идентификаторы для получения дополнительной информации с IMDB, что позволяет обогатить наш анализ реальными производственными данными.

________________
Конструктор класса Links \
Конструктор принимает путь к файлу links.csv и проверяет его существование через os.path.exists(). При успешной проверке сохраняет путь в self.path, затем последовательно вызывает методы для чтения и обработки данных: read_file() загружает первые 1000 строк файла, select_movieId() извлекает movieId (первая колонка), select_imdbId() извлекает imdbId (вторая колонка), select_tmdbId() извлекает tmdbId (третья колонка). Все эти данные сохраняются как атрибуты объекта для последующего использования. Если файл не найден, выбрасывается FileNotFoundError с указанием проблемного пути. Такая структура обеспечивает немедленную загрузку данных при создании объекта и доступ ко всем необходимым идентификаторам через атрибуты класса.

Чтение и парсинг файла (read_file) \
Метод читает файл links.csv, пропускает заголовок и загружает только первые 1000 строк. Каждая строка очищается от пробельных символов с помощью strip(). Результат возвращается как список строк, что обеспечивает дальнейшую обработку данных в других методах класса.

```python
%timeit

links = Links(PATH_TO_CSV_LINKS, PATH_TO_CSV_MOVIES)
sample_movies = links.movieId[:20]
```

____________________
Получение IMDB информации для фильмов (get_imdb) \
Метод получает список фильмов и полей, которые нужно извлечь из IMDB. Для каждого movieId определяется соответствующий imdbId через поиск индекса в списке self.movieId. Затем для каждого поля вызывается метод imdb_parsing(), который парсит данные с веб-страницы. Результат собирается в виде списка списков [movieId, поле1, поле2, ...]. В конце список сортируется по убыванию movieId с использованием lambda-функции, которая преобразует строковый идентификатор в целое число.

Парсинг данных с IMDB (imdb_parsing) \
Этот метод выполняет HTTP-запрос к странице фильма на IMDB с использованием заголовков браузера для обхода защиты. Полученный HTML парсится с помощью BeautifulSoup. Для каждого поддерживаемого поля (Director, Budget, Gross worldwide, Runtime, Title) используется своя логика поиска: сначала находится элемент с соответствующим label, затем переходится к родительскому li и извлекаются нужные данные. Для числовых значений используется регулярное выражение для извлечения чисел и удаления запятых. Для Director и других полей с несколькими значениями собирается список.

```python
%timeit

try:
    imdb_info = links.get_imdb(sample_movies[:3], ['Title', 'Director', 'Budget', 'Runtime'])
    
    for film_info in imdb_info:
        print(f"\nФильм ID: {film_info[0]}")
        print(f"  Название: {film_info[1]}")
        print(f"  Режиссер: {film_info[2]}")
        print(f"  Бюджет: ${film_info[3]:,}" if film_info[3] else "  Бюджет: не указан")
        print(f"  Продолжительность: {film_info[4]} минут" if film_info[4] else "  Продолжительность: не указана")
except Exception as e:
    print(e)
```

_____________
Топ режиссеров по количеству фильмов (top_directors) \
Метод использует get_imdb() для получения информации о режиссерах для списка фильмов. Все найденные режиссеры собираются в один список, включая случаи, когда у фильма несколько режиссеров. Используется Counter для подсчета встречаемости каждого режиссера. Результат сортируется сначала по количеству фильмов (убывание), затем по имени режиссера, и возвращается как словарь, обрезанный до n элементов.

```python
%timeit

try:
    top_directors = links.top_directors(10, sample_movies)

    print("Лучшие режиссеры:")
    for director, film_count in top_directors.items():
        print(f"  {director}: {film_count} фильма(ов)")
except Exception as e:
    print(e)
```

____________
Самые дорогие фильмы по бюджету (most_expensive) \
Метод получает бюджеты фильмов через get_imdb(). Создается словарь с парами "название: бюджет". Этот словарь сортируется по убыванию бюджета, а при одинаковых бюджетах - по алфавиту названий. Возвращаются первые n элементов.

```python
%timeit

try:
    expensive_films = links.most_expensive(5, sample_movies)

    print("Самые дорогие фильмы:")
    for title, budget in expensive_films.items():
        print(f"  '{title}': бюджет ≈ ${budget:,}")
except Exception as e:
    print(e)
```

Самые прибыльные фильмы (most_profitable) \
Метод извлекает мировые сборы и бюджет через get_imdb(). Для каждого фильма вычисляется прибыль как разница между сборами и бюджетом. Создается словарь "название: прибыль", который сортируется по убыванию прибыли, а при равных значениях - по названию. Возвращаются первые n элементов.

```python
%timeit

try:
    profitable_films = links.most_profitable(5, sample_movies)

    print("Самые прибыльные фильмы:")
    for title, profit in profitable_films.items():
        print(f"  '{title}': прибыль ≈ ${profit:,}")
except Exception as e:
    print(e)
```

Самые длинные фильмы (longest) \
Метод получает продолжительность фильмов в минутах через get_imdb(). Создается словарь "название: продолжительность". Сортировка выполняется по убыванию продолжительности, а при равной продолжительности - по алфавиту названий. Возвращаются первые n элементов.

```python
%timeit

try:
    longest_films = links.longest(5, sample_movies)

    print("Самые длинные фильмы:")
    for title, runtime in longest_films.items():
        hours = runtime // 60
        minutes = runtime % 60
        print(f"  '{title}': {hours}ч {minutes}м ({runtime} минут)")
except Exception as e:
    print(e)
```

_______________
Наибольшая стоимость минуты фильма (top_cost_per_minute) \
Метод вычисляет соотношение "бюджет/продолжительность" для каждого фильма. Результат округляется до двух знаков после запятой. Создается словарь "название: стоимость_минуты", который сортируется по убыванию этого показателя, а при равных значениях - по названию. Возвращаются первые n элементов.

```python
%timeit

try:
    cost_per_minute = links.top_cost_per_minute(5, sample_movies)

    print("Стоимость минуты экранного времени:")
    for title, cost in cost_per_minute.items():
        print(f"  '{title}': ${cost:,.2f} за минуту")
except Exception as e:
    print(e)
```
